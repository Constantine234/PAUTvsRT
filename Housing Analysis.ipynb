```python
# Create the bar chart
plt.figure(figsize=(10, 6))  # Optional: set figure size
plt.bar(df,color='skyblue')

# Add title and labels
plt.title('Sample Bar Chart')
plt.xlabel('Location')
plt.ylabel('Price')

# Display the plot
plt.show()
```

```python
# Calculate average area by condition
avg_area_by_condition = df.groupby('Condition')['Area'].mean()

# Plot the bar chart
plt.figure(figsize=(10, 6))
avg_area_by_condition.plot(kind='bar', color='skyblue')
plt.title('Average Area by Condition')
plt.xlabel('Condition')
plt.ylabel('Average Area (sq ft)')
plt.show()
```

### Summary: Average Area by Condition
The Average Area by Condition analysis shows that properties in both Fair and Excellent condition have larger average areas compared to other conditions. This suggests that:

Fair condition properties may include larger, possibly older homes requiring updates.
Excellent condition properties are likely well-maintained or recently renovated, reflecting higher investment in upkeep.
These insights indicate a possible link between property size and maintenance level, with the largest properties either well-kept or needing updates. Further analysis could explore trends by location or property age to deepen this understanding.

```python
# Calculate average price by location
avg_price_by_location = df.groupby('Location')['Price'].mean()

# Plot the bar chart
plt.figure(figsize=(10, 6))
avg_price_by_location.plot(kind='bar', color='lightcoral')
plt.title('Average Price by Location')
plt.xlabel('Location')
plt.ylabel('Average Price')
plt.show()
```

### Average Price by Location Analysis
The Average Price by Location bar chart reveals a key trend:

Suburban Areas: Show the highest average property prices. This is likely due to larger property sizes and a preference for quieter, family-friendly environments, which are appealing to many buyers.

Urban Areas: Display the lowest average prices, possibly due to higher density and smaller property sizes despite the proximity to amenities.

This trend indicates that suburban properties may offer higher market value compared to urban ones, providing valuable insight for potential investment strategies.
